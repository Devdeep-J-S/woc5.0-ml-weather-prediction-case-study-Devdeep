1) Reading the above CSV file in google colab using the pandasâ€™ library and converting it into data frames.

```python
# data from drive 
import pandas as pd
path = 'https://raw.githubusercontent.com/Devdeep-J-S/woc5.0-ml-weather-prediction-case-study-Devdeep/main/WOC5.0-ML-Checkpoint-2/Youtube%20Influencer%20Analysis%20-%20Updated.csv'
df = pd.read_csv(path,sep=",", encoding='mac_roman')
```

2) How many values are missing for each column and try dropping or replacing NULL values using appropriate pandasâ€™ function.

```python
# how many values are missing 
# before 
print("Before dropping rows with missing values")
print(df.isna().sum())

# droping rows with null values 
# method -1 : df[df.isnull().any(axis=1)]
df.dropna(inplace=True)

#after 
print("After dropping rows with missing values")
print(df.isna().sum())
```

3) Drop the columns which are irrelevant features to feed into our model.



```python
# suppose duration of video is irrelevent feature 
df.drop(['Duration of Video'],axis=1,inplace=True)
```

4) Some columns are not statistical data but in the form of strings. 
It Is difficult for our model to learn when the data is in form of strings. 
Your task is to encode these strings using proper encoding techniques.

```python
df.dtypes
# good resources : https://pbpython.com/categorical-encoding.html
# https://www.geeksforgeeks.org/ml-one-hot-encoding-of-datasets-in-python/

# for this question I will be encoding the "Creator gender" and "Language of the video" columns using proper encoding techniques.

# method -1 
temp = pd.get_dummies(df["Creator Gender"])
```

```python
print(temp)
```

```python
df["Male"] = temp["Male"].to_list() 
df["Female"] = temp["Female"].to_list()
df.head()
```

```python
# method 2  
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
label = label_encoder.fit_transform(df["Language of the Video"])
df["Language of the Video"] = label
df.head()
```
