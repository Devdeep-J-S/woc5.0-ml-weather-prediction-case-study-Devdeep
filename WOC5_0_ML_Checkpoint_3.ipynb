```python
import pandas as pd 
car_data = pd.read_csv("/content/drive/MyDrive/CarPrice_Assignment.csv")
car_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# simple plt 
plt.plot(car_data.price);
plt.xlabel("Cars");
plt.ylabel("Car Price");
plt.figure(figsize=(20,10));
```

```python
sns.set_style("darkgrid")
plt.plot(car_data.enginesize,car_data.price,'or');
plt.xlabel("Cars engine size ");
plt.ylabel("Car Price");
plt.title("Graph with dark background")
plt.figure(figsize=(20,10));
```

```python
# scatterplot 
sns.scatterplot(car_data.wheelbase,car_data.price);
```

```python
# hue to add more info 
# s to adjust size of the points 
plt.figure(figsize=(20,10))
sns.scatterplot(car_data.wheelbase,car_data.price,hue=car_data.CarName,s=100);
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(car_data.wheelbase,car_data.price,hue=car_data.fueltype,s=100);
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(car_data.wheelbase,car_data.horsepower,car_data.price);
```
